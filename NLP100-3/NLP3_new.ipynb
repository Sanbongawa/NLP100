ファイル全体はgzipで圧縮される
以下の処理を行うプログラムを作成せよ．

#### 20. JSONデータの読み込み
Wikipedia記事のJSONファイルを読み込み，「イギリス」に関する記事本文を表示せよ．問題21-29では，ここで抽出した記事本文に対して実行せよ．

#### 21. カテゴリ名を含む行を抽出

#### 27. 内部リンクの除去
26の処理に加えて，テンプレートの値からMediaWikiの内部リンクマークアップを除去し，テキストに変換せよ（参考: マークアップ早見表）．

#### 28. MediaWikiマークアップの除去
27の処理に加えて，テンプレートの値からMediaWikiマークアップを可能な限り除去し，国の基本情報を整形せよ．

#### 29. 国旗画像のURLを取得する

# 22. カテゴリ名の抽出

```python
for i in uk:
    a = i.find('Category')
    if a == -1:
        pass
    else:
        b = re.search(r'(?<=:)\S+\w',i)
        print(b.group(0))
```

# 24. ファイル参照の抽出

```python
media=['JPEG','jpg','PNG','png','mp3','MP3','svg','SVG']
for j in media:
    for i in uk:
        a = i.find(j)
        if a == -1:
            pass
        else:
            #print(i)
            e = re.sub(r'[|].*',' ',i)
            e = re.sub(r'[[]','',e)
            print(e)
```

# 25. テンプレートの抽出

```python
info = UK[1]#.replace(' ','')
Info=info.split('|')
w = ''
for item in Info:
    w = w+item
w = w.split('\n')
w
```

```python
from collections import OrderedDict
```

```python
uk_info = OrderedDict()
```

```python
pattan = re.compile(r'(.*)\s\=\s(.*)')#.(*?\=.+\s?')
#result = re.search(pattan,w[4])
for i in w:
    result = pattan.search(i)
    if result:
        if '==' in i:
            break
        else:
            print(result.group(0))
            uk_info[result.group(1)] = result.group(2)
            
```

# 26. 強調マークアップの除去

```python
info = UK[1]#.replace(' ','')
Info=info.split('|')
w = ''
for item in Info:
    w = w+item
w = w.split('\n')
uk_info1 = OrderedDict()
pattan = re.compile(r'(.*)\s\=\s(.*)')#.(*?\=.+\s?')

for i in w:
    i = re.sub(r'\'{2,5}','\'',i)
    result = pattan.search(i)
    if result:
        if '==' in i:
            break
        else:
            print(result.group(0))
            uk_info1[result.group(1)] = result.group(2)
         
```

# 27. 内部リンクの除去

```python
info = UK[1]#.replace(' ','')
Info=info.split('|')
w = ''
for item in Info:
    w = w+item
w = w.split('\n')
uk_info2 = OrderedDict()
pattan = re.compile(r'(.*)\s\=\s(.*)')#.(*?\=.+\s?')

for i in w:
    i = re.sub(r'\'{2,5}','\'',i)
    i = re.sub(r'[\[\]]','',i)
    #i = re.sub(r'(?=\<).*\>','',i)
    if '注記' in i:
        pass
    else:
        i = re.sub(r'(?=\<).*\>','',i)
    result = pattan.search(i)
    if result:
        if '==' in i:
            break
        else:
            print(result.group(0))
            uk_info2[result.group(1)] = result.group(2)
      
```

# 28. MediaWikiマークアップの除去

```python
info = UK[1]#.replace(' ','')
Info=info.split('|')
w = ''
for item in Info:
    w = w+item
w = w.split('\n')
uk_info3 = OrderedDict()
pattan = re.compile(r'(.*)\s\=\s(.*)')#.(*?\=.+\s?')

for i in w:
    i = re.sub(r'\'{2,5}','\'',i)
    i = re.sub(r'[\[\]]','',i)
    #i = re.sub(r'(?=\<).*\>','',i)
    if '注記' in i:
        pass
    else:
        i = re.sub(r'(?=\<).*\>','',i)
    
    i = re.sub(r'[\#\{\}\~\!\*\;\|]','',i)
    result = pattan.search(i)
    if result:
        if '==' in i:
            break
        else:
            print(result.group(0))
            uk_info3[result.group(1)] = result.group(2)
      
```

# 29. 国旗画像のURLを取得する

```python
info = UK[1]#.replace(' ','')
Info=info.split('|')
w = ''
for item in Info:
    w = w+item
w = w.split('\n')
pattan = re.compile(r'(.*)\s\=\s(.*)')#.(*?\=.+\s?')
flag_url=''

for i in w:
    i = re.sub(r'\'{2,5}','\'',i)
    i = re.sub(r'[\[\]]','',i)
    #i = re.sub(r'(?=\<).*\>','',i)
    if '注記' in i:
        pass
    else:
        i = re.sub(r'(?=\<).*\>','',i)
    
    i = re.sub(r'[\#\{\}\~\!\*\;\|]','',i)
    result = pattan.search(i)
    if result:
        if '==' in i:
            break
        else:
            if '国旗' in result.group(0):
                flag_url += re.sub(r' ','_',result.group(2))
                print(flag_url)
            else:
                pass
```

```python
wiki_api = 'http://ja.wikipedia.org/w/api.php?action=query&rawcontinue=True&format=json&\
titles=Image:%s&prop=imageinfo&iiprop=url|dimensions|mime'
```

```python
from urllib import error,request
```

```python
send_url = wiki_api % flag_url
send_url
```

```python
res = request.urlopen(send_url)
data = res.read()
data
```
